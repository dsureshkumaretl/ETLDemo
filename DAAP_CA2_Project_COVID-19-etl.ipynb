```python
# Reading covid19_vaccinations-statistics data #Data Set 1 #1.8 Million Data
 # Data Extraction :
# MongoDB Table 1
covid19_vaccinations_statistics_url = "https://data.cdc.gov/api/views/8xkx-amqh/rows.json?accessType=DOWNLOAD"
covid19_vaccinations_statistics_response = urllib.request.urlopen(covid19_vaccinations_statistics_url)
covid19_vaccinations_statistics_data = json.loads(covid19_vaccinations_statistics_response.read())
print("API request completed status",covid19_vaccinations_statistics_response.status)
# inserting into mongoDB
covid19_vaccinations_collection = project_db["covid19_vaccinations"]
covid19_vaccinations_cols = list(filter(lambda val: val["id"] != -1, covid19_vaccinations_statistics_data["meta"]["view"]["columns"]))
covid19_vaccinations_cols_others_len = len(list(filter(lambda val: val["id"] == -1, covid19_vaccinations_statistics_data["meta"]["view"]["columns"])))

for row in covid19_vaccinations_statistics_data["data"]:
    db_row = {}
    for col in covid19_vaccinations_cols:
        db_row[col["fieldName"]] = row[int(col["position"]) + covid19_vaccinations_cols_others_len - 1]
    covid19_vaccinations_collection.insert_one(db_row)

print("Inserted unemployment_statistics document count: ", covid19_vaccinations_collection.count_documents({}))
```
